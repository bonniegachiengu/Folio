```python
class Wallet:
    """
    A Wallet is unique housing for a user's internal ledger, portfolio and profile.
    The name of the Wallet is the user's username.
    """
    
    def __init__(self, username):
        self.username = username

    def __str__(self):
        return f"{self.username}"
    
    @classmethod
    def get(cls):
        username = input("Choose username? ")
        return cls(username)
    
    @property
    def username(self):
        return self._username
    
    @username.setter
    def username(self, username):
        if not username:
            raise ValueError("Missing username")
        self._username = username

def main():
    wallet = Wallet.get()
    print(wallet)


if __name__ == "__main__":
    main()
```
